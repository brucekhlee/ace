```python
def align_rms(target1, target2):
    cmd.load(target1, "t1")
    cmd.load(target2, "t2")
    cmd.align("t1", "t2")
    _rms = cmd.rms("t1", "t2")
    cmd.delete("all")
    return _rms
```

```python
for i in range(len(testing)):
    print(next(open(testing[i])).replace("\n","")[10:],
          round(align_rms(ref,testing[i]),2),
          testing[i].replace("/home/aguan/docking/docking52tcksXP088/",""))
```
